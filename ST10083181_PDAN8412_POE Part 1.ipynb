# What is a recurrent neural network?
-	According to Awan. A (2022), Recurrent Neural Networks (RNNs) are a specialized type of artificial neural network (ANN) designed to process and analyse sequential data. RNNs excel in tasks like voice recognition in Apple’s Siri and Google’s voice search by leveraging an internal memory that enables them to remember and utilize previous inputs. This memory function is what distinguishes RNNs from traditional neural networks, where each input is treated independently. The ability of RNNs to maintain context over time makes them particularly effective for applications involving time series prediction, text generation, transcription, and machine translation, where the sequence of data points is crucial for accurate prediction or classification.

-	Awan. A (2022), states that RNNs work by introducing a loop within the network structure that allows information to be passed from one step of the sequence to the next, creating a "memory" of previous inputs. Unlike feedforward neural networks, where inputs and outputs are treated independently, RNNs take into account the sequence of inputs. The hidden layer in an RNN functions as this memory, capturing information from all previous time steps. During processing, the network updates the hidden state with each new input, ensuring that the sequence's context is preserved. This allows the network to make predictions or classifications that depend on prior elements in the sequence. The training process in RNNs utilizes a specialized form of backpropagation called backpropagation through time (BPTT), which sums errors across time steps to adjust weights and biases appropriately.

-	RNNs are versatile in their application due to the different lengths of inputs and outputs they can handle. The types of RNNs According to Awan. A (2022), include:
1.	One-to-One: This is the simplest form of RNN, used for problems where there is a single input and a single output, similar to traditional neural networks.
2.	One-to-Many: This type takes a single input and produces multiple outputs. It is often used in tasks like image captioning, where a single image is the input, and a sequence of words (a caption) is the output.
3.	Many-to-One: This type processes a sequence of inputs to produce a single output. A common application is sentiment analysis, where the input is a sequence of words (text), and the output is a sentiment category.
4.	Many-to-Many: This type handles sequences as both input and output, making it ideal for tasks like machine translation, where a sentence in one language is translated into a sentence in another language.

-	The structure of an RNN is designed to incorporate sequential memory by introducing loops that allow information to persist within the network. Unlike traditional feedforward networks, which have a linear flow of data from input to output, RNNs feature a looping mechanism that acts as a conduit for information to flow from one time step to the next. This mechanism enables the hidden state to function as the network's memory, capturing and retaining information from all previous time steps. During processing, each word or data point is transformed into a machine-readable vector, and the RNN processes these vectors sequentially. The hidden state is updated continuously, combining the current input with the previous hidden state to form a new vector. This vector is passed through an activation function like tanh, which ensures that the values remain within a certain range, preventing the outputs from exploding or vanishing. This process allows RNNs to maintain a memory of prior inputs while processing new data, making them capable of handling sequential information effectively.

-	According to Anon (2024), RNNs are based on the recursive application of a function to a sequence of inputs. Mathematically, the hidden state ℎ𝑡 at time step 𝑡 is defined as ℎ𝑡=𝑓(𝑊ℎℎ𝑡−1+𝑊𝑥𝑥𝑡+𝑏), where 𝑊ℎ and 𝑊𝑥 are weight matrices for the hidden state and the input, respectively, 𝑏 is the bias, and 𝑓 is an activation function (often tanh or ReLU). The output 𝑦𝑡 is typically calculated as 𝑦𝑡=𝑔(𝑊𝑦ℎ𝑡+𝑐) where, 𝑊𝑦 is the weight matrix for the output and 𝑐 is the bias. 

-	According to Anon (2024), Training RNNs involves Backpropagation Through Time (BPTT), a specialized form of backpropagation that accounts for errors across all time steps. This process can be computationally intensive due to the need to unroll the network and compute gradients for each time step. Challenges such as vanishing and exploding gradients can arise, which can be mitigated using techniques like gradient clipping and advanced architectures like LSTM and GRU.

-	Anon (2024), states that to reduce the computational load, a variant called Truncated BPTT is often used, where the sequence is split into smaller chunks, and backpropagation is applied only over these chunks rather than the entire sequence. This reduces the memory requirement and speeds up training, albeit at the cost of potentially losing some long-term dependencies.

-	Anon (2024), states that to prevent overfitting, regularization techniques such as dropout (applied between RNN layers) are used during training. Dropout randomly disables a fraction of the neurons during each training iteration, which helps to prevent the network from becoming too dependent on specific neurons and thus improves generalization.

-	According to Awan. A (2022) RNNs offer several advantages, particularly in handling sequential data where the order of inputs is critical. However, they also come with notable limitations. One major challenge is the vanishing gradient problem, where the gradients used to update the network’s weights become so small that learning effectively stops, leading to poor performance in long sequences. Conversely, the exploding gradient problem occurs when gradients become too large, causing instability in the model. These issues can be mitigated with advanced architectures like LSTM and GRU, which are designed to handle long-term dependencies and maintain stable learning over time. Despite these challenges, RNNs remain a powerful tool for tasks that require an understanding of sequential patterns, offering unique advantages in fields such as natural language processing, time series forecasting, and speech recognition.

-	To address some of the inherent limitations of standard RNNs, advanced architectures such as Long Short-Term Memory (LSTM) and Gated Recurrent Unit (GRU) were developed. According to Awan. A (2022), These architectures introduce gating mechanisms that regulate the flow of information through the network, allowing it to retain relevant information over longer periods and discard irrelevant data. LSTMs, for instance, have a more complex structure with multiple interacting layers that work together to maintain long-term dependencies within the sequence. GRUs, while simpler in structure, achieve similar results by using update and reset gates to control the hidden state. These advanced architectures are particularly useful in tasks that require the network to remember information from far back in the sequence, such as speech recognition, text generation, and machine translation.









# Why is the chosen dataset appropriate for analysis with an RNN?
- This dataset comprises 100,000 reviews of the ChatGPT app, sourced from the Google Play Store. It includes diverse feedback from users in ten countries, offering valuable insights into user experiences and app performance. The dataset is ideal for natural language processing tasks, sentiment analysis, and user feedback studies.

1. Sequential Nature of Review Text:
-	RNNs are specifically designed to process sequential data by maintaining a hidden state that evolves over time. This ability to capture temporal dependencies makes RNNs ideal for analysing text where the sequence of words impacts the meaning. According to Awan, A (2022), RNNs use loops to pass information from one time step to the next, making them well-suited for tasks like sentiment analysis and text classification, which are central to this ChatGPT reviews dataset. The dataset includes comments, which are sequential in nature. Each review is a sequence of words where the order of words is crucial for understanding the sentiment and context. The ‘comment’ feature contains strings of varying lengths, each representing a sequence of words. A comment may change its meaning entirely based on word order, making RNNs particularly well-suited for this type of sequential processing.

2. Handling Long-Term Dependencies:
-	RNNs, and more advanced variants like LSTM and GRU, can maintain and utilize contextual information across long sequences. LSTMs, for example, use gating mechanisms to manage long-term dependencies and retain relevant information, addressing the issue of vanishing gradients as highlighted by Hochreiter, S and Schmidhuber, J (1997). This capability is crucial for accurately interpreting user reviews where context plays a significant role. Comments on a review often involve nuanced opinions and feedback that are dependent on the context of preceding words. LSTMs are specifically designed to manage the issue of vanishing gradients, allowing them to retain important information over longer sequences of text. This capability is essential for processing lengthy reviews in the ChatGPT dataset, where key sentiment cues might appear early on and must be retained through the end of the review.

3. Handling Variable-Length Sequences:
-	RNNs process sequences iteratively, which allows them to handle varying lengths of input data. This flexibility is a key advantage of RNNs over traditional neural networks that often require fixed-size inputs. This adaptability is emphasized in sequence modelling research, such as that by Bengio, Y, Simard, P et al, (1994), which demonstrates how RNNs can manage sequences of arbitrary lengths. The length of reviews in this dataset varies widely, from short comments to detailed feedback. RNNs are capable of handling inputs of variable lengths without requiring a fixed size, making them suitable for this dataset.

4. Backpropagation Through Time (BPTT):
- BPTT is designed to handle the unique challenges of sequential data by calculating gradients across all time steps. This approach allows RNNs to learn from the context provided by the entire sequence, addressing the temporal dependencies in reviews in the dataset. As discussed by Hinton, G. et al. (1986), BPTT enables the RNN to adjust weights based on accumulated errors over time, enhancing its ability to process and understand complex sequences. Training the RNN to understand and generate text involves using Backpropagation Through Time (BPTT), a specialized training method for sequential data. This method helps in learning from the entire sequence of words in each review. In the ChatGPT Reviews dataset, BPTT enables the model to effectively process and understand complex sentences by adjusting weights based on accumulated errors across the sequence. This is particularly important for understanding the nuances in user reviews where the sentiment or meaning may depend on multiple words spread across a long text.

5. Sentiment Prediction and Label Mapping:
- RNNs are highly effective at mapping sequences of text to corresponding labels, a process crucial for tasks like sentiment analysis or predicting ratings. According to Sutskever, I. et al. (2014), sequence-to-sequence models built on RNNs can be used for predictive tasks involving sequential data. This functionality makes RNNs an ideal choice for analysing user feedback in this dataset, where the goal is to predict or classify user sentiment based on review text. In the ChatGPT dataset, where reviews are paired with ratings and thumbs up/down counts, RNNs can learn to predict the overall sentiment or rating based on the review's textual content or comments.

6. Integration with Categorical Data:
- RNNs can be integrated with other types of data (e.g., categorical variables) by combining them in a hybrid model. This approach leverages the sequential processing capabilities of RNNs along with other feature inputs to improve overall model performance. This integration is supported by advancements in hybrid neural network architectures that combine sequential and non-sequential data (Zhang, X. et al., 2016). Beyond textual data, the dataset also includes categorical variables like app version, country, and rating. This integration allows for a more robust model capable of making nuanced predictions based on a combination of textual and non-textual data.

## 3.	Dealing with Overfitting and Underfitting:

**Overfitting** 
-	Dropout: Apply dropout layers in your RNN/LSTM architecture. Dropout randomly disables a fraction of neurons during training to prevent the model from becoming too reliant on specific neurons. This helps in generalizing better. Common dropout rates are between 0.2 and 0.5.
-	L2 Regularization: Add L2 regularization (weight decay) to the loss function. This penalizes large weights and helps in keeping the model weights smaller, thus preventing overfitting.

**Underfitting**
-	Feature Engineering: Provide the model with additional relevant features to help it learn better. Extract and incorporate additional features from the text data, such as sentiment scores or named entities. 
-	Tune Hyperparameters: Optimize hyperparameters to improve the model’s capacity to learn from the data. Experiment with different values for learning rate, number of hidden units, batch size, and sequence length. Use techniques like random search to find optimal values.

## 5.	Hyperparameter Tuning:

-	 Number of LSTM Units/Cells: Determines the capacity of the LSTM network. More units can capture more complex patterns but may lead to overfitting.
-	Number of Layers: The depth of the network. More layers can capture more complex features but might increase training time and risk overfitting.
-	Dropout Rate: A regularization technique to prevent overfitting. Dropout randomly sets a fraction of input units to 0 at each update during training time.
-	Learning Rate: The step size for gradient descent. A too high rate may cause the model to converge too quickly to a suboptimal solution, while a too low rate may result in a longer training time.
-	Batch Size: The number of samples processed before the model is updated. Larger batch sizes can speed up training but may require more memory.
 Epochs: The number of times the entire training dataset is passed through the network. More epochs can lead to better training but may also cause overfitting.

-	Strategy - Random Search: Randomly samples combinations of hyperparameters from the search space. This is less exhaustive but often more efficient than grid search.

## 6. Interpreting Model Results: 

1. Interpreting Performance Scores
-	Accuracy: Measures the proportion of correctly classified instances among the total instances. High accuracy indicates that the model is correctly predicting a large portion of the reviews. However, accuracy alone may not be sufficient if the dataset is imbalanced.
-	Precision: The ratio of true positives to the sum of true positives and false positives. High precision indicates that when the model predicts a positive class, it is correct most of the time.
-	Recall: The ratio of true positives to the sum of true positives and false negatives. High recall indicates that the model identifies most of the positive instances.
-	F1-Score: The harmonic mean of precision and recall. It provides a single metric that balances precision and recall, useful for imbalanced datasets.

2. Analysing the Confusion Matrix
The confusion matrix provides a detailed breakdown of the model’s performance:
-	True Positives (TP): Correctly predicted positive reviews.
-	True Negatives (TN): Correctly predicted negative reviews.
-	False Positives (FP): Negative reviews incorrectly predicted as positive.
-	False Negatives (FN): Positive reviews incorrectly predicted as negative.
-    From the confusion matrix, we can derive:
     -	Class Imbalance: If the matrix shows that some classes have many more instances than others, the model might be biased towards the majority class. This can be addressed by techniques such as resampling or adjusting class weights.
     -	Error Types: Identify which types of errors are most common (e.g., the model frequently misclassifies positive reviews as negative). This helps in understanding where the model’s predictions are falling short and guides potential improvements.

3. Effects of Hyperparameter Tuning
-	Impact on Performance Metrics: After tuning hyperparameters, compare the performance metrics (accuracy, precision, recall, F1-score) with those obtained before tuning. Improved scores indicate successful tuning. For example:
-	Learning Rate: A well-tuned learning rate should lead to faster convergence and better accuracy. Too high a rate might result in unstable training, while too low might slow down convergence.
-	Number of LSTM Units/Layers: Increasing units or layers can capture more complex patterns but may lead to overfitting if not properly regularized.
-	Dropout Rate: Adjusting the dropout rate can help mitigate overfitting. A high dropout rate might improve generalization but could also lead to underfitting if too much information is discarded.
-	Visualization of Tuning Effects: Plot learning curves (training and validation loss/accuracy) for different hyperparameter settings to visually assess improvements and identify overfitting or underfitting issues.

```python
# Developing RNN and applying LSTM
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from sklearn.metrics import accuracy_score
import itertools

# Define device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Sample and split data
sampled_df = data_cleaned.sample(fraction=0.2, seed=42)
train_df, test_df = sampled_df.randomSplit([0.8, 0.2], seed=42)


# Define number of features and classes
num_features = 59879  # Set this to the actual number of features in your dataset
num_classes = len(sampled_df.select("Rating").distinct().collect())  # Number of unique classes

# Define criterion for loss function
criterion = nn.CrossEntropyLoss()

# Define dataset class
class SparseDataset(Dataset):
    def __init__(self, df):
        features_list = [row["datasetFeatures"] for row in df.select("datasetFeatures").collect()]
        self.features = [torch.tensor(vec.toArray(), dtype=torch.float) for vec in features_list]
        labels_list = [row["Rating"] for row in df.select("Rating").collect()]
        
        # Ensure labels are within the valid range
        self.labels = torch.tensor([min(label, num_classes - 1) for label in labels_list], dtype=torch.long)
    
    def __len__(self):
        return len(self.labels)
    
    def __getitem__(self, idx):
        return self.features[idx], self.labels[idx]

# Define collate_fn function
def collate_fn(batch):
    features, labels = zip(*batch)
    features = torch.stack(features, dim=0)
    labels = torch.tensor(labels, dtype=torch.long)
    return features, labels

# Create Dataloaders
train_dataset = SparseDataset(train_df)
test_dataset = SparseDataset(test_df)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, collate_fn=collate_fn)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False, collate_fn=collate_fn)

# Define models
class RNNModel(nn.Module):
    def __init__(self, input_size, hidden_size, num_classes, num_layers=1, dropout=0.5):
        super(RNNModel, self).__init__()
        self.rnn = nn.RNN(input_size, hidden_size, num_layers, dropout=dropout, batch_first=True)
        self.fc = nn.Linear(hidden_size, num_classes)
    
    def forward(self, x):
        out, _ = self.rnn(x)
        out = self.fc(out[:, -1, :])  # Only take the output of the last time step
        return out

class LSTMModel(nn.Module):
    def __init__(self, input_size, hidden_size, num_classes, num_layers=1, dropout=0.5):
        super(LSTMModel, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, dropout=dropout, batch_first=True)
        self.fc = nn.Linear(hidden_size, num_classes)
    
    def forward(self, x):
        out, _ = self.lstm(x)
        out = self.fc(out[:, -1, :])  # Only take the output of the last time step
        return out

# Define hyperparameter space
hidden_sizes = [50, 100, 150]
num_layers = [1, 2]
learning_rates = [0.001, 0.01]
dropout_rates = [0.2, 0.5]

# Generate all combinations of hyperparameters
hyperparameter_space = list(itertools.product(hidden_sizes, num_layers, learning_rates, dropout_rates))

# Training function
def train_model(model, train_loader, optimizer, num_epochs=5):
    model.train()
    train_losses = []
    early_stopping = EarlyStopping(patience=5, delta=0.01)  # Initialize early stopping
    for epoch in range(num_epochs):
        epoch_loss = 0
        try:
            for features, labels in train_loader:
                features, labels = features.to(device), labels.to(device)
                outputs = model(features.unsqueeze(1))  # Add sequence dimension
                loss = criterion(outputs, labels)
                optimizer.zero_grad()
                loss.backward()
                optimizer.step()
                epoch_loss += loss.item()

            print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {epoch_loss / len(train_loader)}")
            val_loss = evaluate_model(model, test_loader)  # Validation loss calculation here
            early_stopping(val_loss)
            if early_stopping.early_stop:
                print("Early stopping triggered.")
                break

            train_losses.append(epoch_loss / len(train_loader))
        except Exception as e:
            print(f"Error during training: {e}")

# Evaluation function
def evaluate_model(model, test_loader):
    model.eval()
    all_labels = []
    all_preds = []
    try:
        with torch.no_grad():
            for features, labels in test_loader:
                features, labels = features.to(device), labels.to(device)
                outputs = model(features.unsqueeze(1))  # Add sequence dimension
                _, predicted = torch.max(outputs.data, 1)
                all_labels.extend(labels.cpu().numpy())
                all_preds.extend(predicted.cpu().numpy())
        accuracy = accuracy_score(all_labels, all_preds)
        return accuracy
    except Exception as e:
        print(f"Error during evaluation: {e}")
        return 0.0

# Early stopping class
class EarlyStopping:
    def __init__(self, patience=5, delta=0):
        self.patience = patience
        self.delta = delta
        self.best_loss = float('inf')
        self.counter = 0
        self.early_stop = False

    def __call__(self, val_loss):
        if self.best_loss - val_loss > self.delta:
            self.best_loss = val_loss
            self.counter = 0
        else:
            self.counter += 1
            if self.counter >= self.patience:
                self.early_stop = True

import random

# Randomly sample 5 combinations of hyperparameters
random_combinations = random.sample(hyperparameter_space, 5)

# Random Search for hyperparameter tuning (limited to 5 combinations)
def random_search(train_loader, test_loader, num_epochs=5):
    best_model = None
    best_accuracy = 0.0
    best_hyperparameters = None

    for hidden_size, num_layer, learning_rate, dropout_rate in random_combinations:
        print(f"Testing hyperparameters: hidden_size={hidden_size}, num_layers={num_layer}, learning_rate={learning_rate}, dropout_rate={dropout_rate}")

        # Initialize models with current hyperparameters
        rnn_model = RNNModel(num_features, hidden_size, num_classes, num_layers=num_layer, dropout=dropout_rate).to(device)
        lstm_model = LSTMModel(num_features, hidden_size, num_classes, num_layers=num_layer, dropout=dropout_rate).to(device)

        # Define loss function and optimizer
        criterion = nn.CrossEntropyLoss()
        rnn_optimizer = optim.Adam(rnn_model.parameters(), lr=learning_rate)
        lstm_optimizer = optim.Adam(lstm_model.parameters(), lr=learning_rate)

        # Train and evaluate RNN model
        try:
            print("Training RNN model...")
            train_model(rnn_model, train_loader, rnn_optimizer, num_epochs=num_epochs)
            print("Evaluating RNN model...")
            rnn_accuracy = evaluate_model(rnn_model, test_loader)
            print(f"RNN Model Accuracy: {rnn_accuracy:.2f}")

            # Update best model
            if rnn_accuracy > best_accuracy:
                best_accuracy = rnn_accuracy
                best_model = rnn_model
                best_hyperparameters = {
                    'hidden_size': hidden_size,
                    'num_layers': num_layer,
                    'learning_rate': learning_rate,
                    'dropout_rate': dropout_rate
                }
        except Exception as e:
            print(f"Error with RNN model: {e}")

        # Train and evaluate LSTM model
        try:
            print("Training LSTM model...")
            train_model(lstm_model, train_loader, lstm_optimizer, num_epochs=num_epochs)
            print("Evaluating LSTM model...")
            lstm_accuracy = evaluate_model(lstm_model, test_loader)
            print(f"LSTM Model Accuracy: {lstm_accuracy:.2f}")

            # Update best model
            if lstm_accuracy > best_accuracy:
                best_accuracy = lstm_accuracy
                best_model = lstm_model
                best_hyperparameters = {
                    'hidden_size': hidden_size,
                    'num_layers': num_layer,
                    'learning_rate': learning_rate,
                    'dropout_rate': dropout_rate
                }
        except Exception as e:
            print(f"Error with LSTM model: {e}")

    print(f"Best model accuracy: {best_accuracy:.2f}")
    print(f"Best hyperparameters: {best_hyperparameters}")

# Run hyperparameter tuning (with 5 combinations)
random_search(train_loader, test_loader, num_epochs=5)


#code attribution
# Ahmed, R. 2020. Hyper-parameter tuning in machine learning models at scale using Pyspark for free. Medium. [Online] Available at: https://medium.com/analytics-vidhya/fine-tuning-at-scale-for-free-2a5c40eedaa2 [Accessed 04 September 2024]
# Imgaboy. 2022. Spark with LSTM. DataBricks Community. [Online] Available at: https://community.databricks.com/t5/machine-learning/spark-with-lstm/td-p/26002  [Accessed 04 September 2024]
# jalFaizy. 2024. Build a Recurrent Neural Network from Scratch in Python. Analytics Vidhya. [Online] Available at: https://www.analyticsvidhya.com/blog/2019/01/fundamentals-deep-learning-recurrent-neural-networks-scratch-python/ [Accessed 04 September 2024]
```

```python
# Learning curves 
import torch
import matplotlib.pyplot as plt

def train_model(model, train_loader, val_loader, criterion, optimizer, num_epochs):
    train_losses = []
    val_losses = []
    train_accuracies = []
    val_accuracies = []
    
    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0
        correct_train = 0
        total_train = 0
        
        for inputs, labels in train_loader:
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            
            running_loss += loss.item()
            _, predicted = torch.max(outputs, 1)
            total_train += labels.size(0)
            correct_train += (predicted == labels).sum().item()
        
        epoch_train_loss = running_loss / len(train_loader)
        epoch_train_accuracy = correct_train / total_train
        
        model.eval()
        running_val_loss = 0.0
        correct_val = 0
        total_val = 0
        
        with torch.no_grad():
            for inputs, labels in val_loader:
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                
                running_val_loss += loss.item()
                _, predicted = torch.max(outputs, 1)
                total_val += labels.size(0)
                correct_val += (predicted == labels).sum().item()
        
        epoch_val_loss = running_val_loss / len(val_loader)
        epoch_val_accuracy = correct_val / total_val
        
        train_losses.append(epoch_train_loss)
        val_losses.append(epoch_val_loss)
        train_accuracies.append(epoch_train_accuracy)
        val_accuracies.append(epoch_val_accuracy)
        
        print(f'Epoch {epoch+1}/{num_epochs}')
        print(f'Train Loss: {epoch_train_loss:.4f}, Train Accuracy: {epoch_train_accuracy:.4f}')
        print(f'Val Loss: {epoch_val_loss:.4f}, Val Accuracy: {epoch_val_accuracy:.4f}')

    return train_losses, val_losses, train_accuracies, val_accuracies

# Call the function
train_losses, val_losses, train_accuracies, val_accuracies = train_model(model, train_loader, val_loader, criterion, optimizer, num_epochs)
```

# Learning curve graphs:
1. Training and Validation Loss Graph:
- X-axis (Epochs): Represents the number of complete passes through the entire training dataset. In this graph, it ranges from 1 to 10.
- Y-axis (Loss): Indicates the loss value, which measures how well the model's predictions match the actual outcomes. Lower loss values generally indicate better performance.
- Blue Line (Training Loss): Shows how the loss on the training dataset decreases over epochs. It typically decreases as the model learns from the data.
- Red Line (Validation Loss): Displays the loss on the validation dataset. This helps in assessing how well the model generalizes to unseen data.
- Explanation:
     - Initial Decrease: The training loss decreases rapidly at first, indicating that the model is learning from the training data.
     - Validation Loss Behavior: Initially decreases but may start to increase or plateau, which can indicate overfitting if it rises while training loss continues to decrease. In this graph, it slightly fluctuates, which could suggest minor overfitting or noise in validation data.
     - Goal: Ideally, both training and validation loss should decrease, indicating the model is learning effectively and generalizing well.

2. Training and Validation Accuracy Graph:
-  X-axis (Epochs): Similar to the loss graph, representing the number of epochs.
- Y-axis (Accuracy): Reflects the proportion of correct predictions made by the model. Higher accuracy indicates better performance.
- Blue Line (Training Accuracy): Shows the model's accuracy on the training dataset over epochs.
- Red Line (Validation Accuracy): Indicates the model's accuracy on the validation dataset.
- Explanation:
     - Initial Increase: Both training and validation accuracy increase rapidly, showing that the model is learning to make better predictions.
     - Training vs Validation Accuracy: If training accuracy continues to increase while validation accuracy plateaus or decreases, it may suggest overfitting. In this graph, validation accuracy reaches a plateau, indicating that the model might be at its generalization limit for the given data.
     - Goal: Both accuracies should ideally increase and converge, indicating effective learning and good generalization.

- These graphs are essential for diagnosing the learning process of RNNs/LSTMs and help in adjusting hyperparameters or employing techniques like regularization to improve model performance.

```python
# Error Analysis
def collect_misclassified_samples(model, data_loader, num_samples=5):
    model.eval()
    misclassified_samples = []

    with torch.no_grad():
        for inputs, labels in data_loader:
            outputs = model(inputs)
            _, predicted = torch.max(outputs, 1)
            incorrect_indices = (predicted != labels).nonzero(as_tuple=True)[0]
            
            for idx in incorrect_indices:
                if len(misclassified_samples) >= num_samples:
                    break
                misclassified_samples.append((inputs[idx], labels[idx], predicted[idx]))
            if len(misclassified_samples) >= num_samples:
                break
    
    return misclassified_samples

misclassified_samples = collect_misclassified_samples(model, val_loader)

def plot_misclassified_samples(misclassified_samples):
    plt.figure(figsize=(12, 6))
    num_samples = len(misclassified_samples)
    
    for i, (input_image, true_label, pred_label) in enumerate(misclassified_samples):
        plt.subplot(1, num_samples, i + 1)
        plt.imshow(input_image.squeeze().cpu(), cmap='gray')  # Adjust this based on your data format
        plt.title(f'True: {true_label.item()}\nPred: {pred_label.item()}')
        plt.axis('off')
    
    plt.show()

plot_misclassified_samples(misclassified_samples)
```
